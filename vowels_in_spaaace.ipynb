## Introduction

A discussion on Twitter between the authors prompted the question: how many a’s should there be in the word “space” when you need to suitably impress someone how big and generally awesome space really is?
“Spaace” is clearly a misspelling, “spaaace” is starting to emphasise it, but how much is too much? And is there any reasonable limit to the number of a’s you can add in space for emphasis? We decided to investigate.


## Methods
We turned to the arbiter of all things popular, the search engine Google.
We searched for the number of hits for the word “space”, “spaace”, “spaaace” and so on, adding one a at a time and noting down the number of hits.
The investigation was interrupted briefly by one of the authour’s partners commenting “What are you doing with your laptop at the breakfast table? Your daughter is staring at you!” but the investigation proceeded shortly afterwards.
The first investigation revealed that there seemed to be a power law distribution with the occasional spike at significant numbers of a’s.

The number of a’s was increased one at a time until the authour got desperate and then started jumping up in steps of five in order to find if there were any zero hits. When this was found at a=125, stepping back to fill in the remaining gaps on 2017 May 15 at 0900.

A search up to a=140 resulted in 0 hits, at which point the first authour felt slightly ridiculous and then stopped.

## Results

The distribution is shown in the figure below.



```python
from bokeh.io import show, output_notebook
from bokeh.plotting import figure
output_notebook()

import pandas
df = pandas.read_csv('vowels_in_space.tsv', sep='\t')

plot_log = figure(width=800, height=400, y_axis_type="log", y_range = [10 ** 0, 10 ** 6], 
                  title="a's in spaaaace as seen by Google in May 2017",
                 x_axis_label = "Number of a's in space",
                 y_axis_label= "Number of hits on Google") 
plot_log.vbar(x=df.na, width=0.8, bottom=0.01, top=df.nhit)\

plot_log.xgrid.grid_line_color = None
plot_log.x_range.start = 0
plot_log.x_range.end = 130
plot_log.legend.orientation = "horizontal"
plot_log.legend.location = "top_center"


show(plot_log)
```

The mode of the distribution is at a=4 and a=5, showing that ‘spaaaace’ and ‘spaaaaace’ provide the majority of emphasis for most people’s purposes. The distribution follows a smooth decline to a=28, then jumps up at a=31 and continues down on a second power law to a=125, where no more hits are reported. This corresponds to an apparent Google limit of 128 characters for a searchable word.
